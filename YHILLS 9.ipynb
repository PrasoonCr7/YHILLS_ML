#Independent        # Dependent 
  Variable             Variable
                                                   Question - Calculate the sales in the 7th week
 x(week)              y(sales in thousand)
 1                    1.2
 2                    1.8
 3                    2.6
 4                    3.2
 5                    3.8

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([1.2,1.8,2.6,3.2,3.8])

model = LinearRegression()
model.fit(x,y)

x_new = np.array([[7]])
y_pred = model.predict(x_new)

print(f'Predicted sales for the 7th week :{y_pred[0]:.2f} thousands')

plt.scatter (x,y,color = 'blue' ,label = 'Actual Sales')
plt.plot(x,model.predict(x),color ='red' ,label = 'Regression Line')
plt.scatter(x_new,y_pred,color = 'green',label = 'Predicted sales for 7th week')
plt.xlabel('week')
plt.ylabel('Sales(in thousand)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

x = np.array([29,15,33,28,39]).reshape(-1,1)
y = np.array([0,0,1,1,1])

model = LogisticRegression()
model.fit(x,y)

x_new = np.array([[33]])
y_prob = model.predict_proba(x_new)

print(f'Probability of passing for the student who studied 33 hours: {y_prob[0][1]:.2f}')
plt.scatter(x,y,color = 'blue' ,label = "Actual data")

x_values = np.linspace(10,45,100).reshape(-1,1)
y_values = model.predict_proba(x_values)[:,1]

plt.plot(x_values , y_values , color = 'red' , label = "logistic Regression Curve")
plt.scatter(x_new , y_prob[0][1] , color = 'green' ,label = "probability for 33 hours")

plt.xlabel('Hours Studied')
plt.ylabel("probability of passing")
plt.legend()
plt.show()
```
