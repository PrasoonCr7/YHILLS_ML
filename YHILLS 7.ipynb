```python
import matplotlib.pyplot as plt 
import numpy as np

# plot 1:
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)  # [ 1 - row , 2 - column , 1st graph ]
plt.plot(x,y)


# plot 2:
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)  # [ 1 - row , 2 - column , 2nd graph ]
plt.plot(x,y)


plt.show() 
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([3,8,1,10,5,7])
y = np.array([77,55,45,53,63,75])

plt.scatter(x , y ,color = 'red')  # in scatter it will give the dots.
plt.show 
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([3,8,1,10,5,7])
y = np.array([77,55,45,53,63,75])

plt.bar(x , y ,color = 'blue') # In bar , it will give the bar graph.
plt.show 
```
