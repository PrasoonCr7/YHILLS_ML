Pandas series a pandas series is like a column in a table. It is a one dimensional holding Data of any type.

```python
import pandas as pd    # s.no - labels
a = [2,5,7]
myvar = pd.Series(a)
print(myvar)
```

```python
print(myvar[0])
```

```python
a = [2,5,7]
myvar = pd.Series(a, index = ['x','y','z'])
print(myvar)
```

```python
print(myvar['y'])
```

** Data Cleaning **
Empty cells
Data in wrong formate
Duplicates

```python
import pandas as pd
df = pd.read_csv("https://cdn.wsform.com/wp-content/uploads/2020/06/industry_sic.csv")
new_df = df.dropna()
print(new_df.to_string())
```

```python
import pandas as pd
df = pd.read_csv("https://cdn.wsform.com/wp-content/uploads/2020/06/industry_sic.csv")
df.fillna(130,inplace = True)
```

```python
print(df.to_string)
```

```python
print(df.duplicated())
```

```python
df.drop_duplicates(inplace = True)
```

```python
print(df.to_string())
```
